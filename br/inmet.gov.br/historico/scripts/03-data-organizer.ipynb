```python
import os
import re
import codecs
from datetime import datetime
```

```python
def header_to_csv(**kwargs):
    """
    Le o arquivo CSV de entrada e converte os headers para colunas do arquivo de saida CSV.
    
    Args:
        in_data_path(str): Caminho completo do arquivo de entrada.
        out_data_dir(str): Caminho da pasta para armazenamento do arquivo de saida.
        data_file(str): Nome do arquivo de entrada.
        delimiter(str): Delimitador de campos utilizado no arquivo de saida
        file_ext(str): Extensao do arquivo de saida
        encoding(str): Codificacao do arquivo de saida
    """
    
    # Log: Mensagem de inicio da conversao
    print(str(datetime.now()) + ': Conversao do arquivo ' + kwargs['data_file'] + ' iniciada.')
    
    # Prepara expressao para capturar os valores dos headers
    header_value_re = re.compile(r'(.+):;(.+)')
    
    # Le o arquivo de entrada
    with codecs.open(kwargs['in_data_path'], 'r', encoding = kwargs['encoding']) as reader:
        
        # Prepara os headers do arquivo para conversao em colunas
        sg_regiao = re.search(header_value_re, reader.readline()).group(2)
        sg_uf = re.search(header_value_re, reader.readline()).group(2)
        nm_estacao = re.search(header_value_re, reader.readline()).group(2)
        cd_estacao_wmo = re.search(header_value_re, reader.readline()).group(2)
        vl_latitude = re.search(header_value_re, reader.readline()).group(2)
        vl_longitude = re.search(header_value_re, reader.readline()).group(2)
        vl_altitude = re.search(header_value_re, reader.readline()).group(2)
        dt_fundacao = re.search(header_value_re, reader.readline()).group(2)
        
        # Leitura dummy de linha para chegar na linha de dados
        header = reader.readline()
        
        # Define os valores das colunas que serao incluidas em cada linha
        header_columns =            sg_regiao      \
            + kwargs['delimiter'] + sg_uf          \
            + kwargs['delimiter'] + nm_estacao     \
            + kwargs['delimiter'] + cd_estacao_wmo \
            + kwargs['delimiter'] + vl_latitude    \
            + kwargs['delimiter'] + vl_longitude   \
            + kwargs['delimiter'] + vl_altitude    \
            + kwargs['delimiter'] + dt_fundacao    \
            + kwargs['delimiter']
        
        # Define o header para o arquivo de saida
        header = 'sg_regiao;sg_uf;nm_estacao;cd_estacao_wmo;vl_latitude;vl_longitude;vl_altitude;dt_fundacao;dt_medicao;hr_medicao;vl_precipitacao_mm;vl_pressao_atmosferica_mb;vl_pressao_atmosferica_mb_max_aut;vl_pressao_atmosferica_mb_min_aut;vl_radiacao_global_kj_m2;vl_temperatura_ar_c_bulbo_seco;vl_temperatura_orvalho_c;vl_temperatura_c_max_aut;vl_temperatura_c_min_aut;vl_temperatura_orvalho_c_max_aut;vl_temperatura_orvalho_c_min_aut;vl_umidade_relativa_max_aut;vl_umidade_relativa_min_aut;vl_umidade_relativa;vl_vento_direcao_horaria_gr;vl_vento_rajada_max_ms;vl_vento_velocidade_ms;vl_none\n'
        
        # Prepara o arquivo de saida
        with codecs.open(kwargs['out_data_path'], 'w', encoding = kwargs['encoding']) as writer:
            
            # Escreve o cabecalho no arquivo de saida
            writer.write(header)
            
            # Para cada linha no arquivo de entrada
            for line in reader:
                writer.write(header_columns + line)
                    
    # Log: Resultado da verificacao dos registros
    print(str(datetime.now()) + ': Conversao do arquivo CSV finalizada.')
```

```python
def organize_files():
    """
    Para cada arquivo de origem, executa a reorganizacao do arquivo CSV.        
    """
    
    # Prepara o dicionario de variaveis (kwargs = keyworded arguments)
    kwargs = {}
    
    # Diretorio de armazenamento dos arquivos originais
    kwargs['in_data_dir'] = '..\\data\\02-cleaned'
    
    # Diretorio de armazenamento dos arquivos tratados
    kwargs['out_data_dir'] = '..\\data\\03-organized'
    
    # Lista dos arquivos originais 
    kwargs['data_files'] = os.listdir(kwargs['in_data_dir'])
    
    # Parametros de organizacao dos arquivos delimitados
    kwargs['delimiter'] = ';'
    kwargs['file_ext'] = '.csv'
    kwargs['encoding'] = 'utf-8'
    
    # Log: Mensagem de inicio do processo 
    print(str(datetime.now()) + ': Organizacao dos arquivos iniciada.')
    
    # Para cada arquivo na lista de arquivos originais
    for file in kwargs['data_files']:
        
        # Define o caminho completo de acesso e escrita dos arquivos
        kwargs['data_file'] = file
        kwargs['in_data_path'] = os.path.join(kwargs['in_data_dir'], kwargs['data_file'])
        kwargs['out_data_path'] = os.path.join(kwargs['out_data_dir'], kwargs['data_file'])
        
        # Corrige registros com menos colunas que o esperado
        header_to_csv(**kwargs)
      
    # Log: Mensagem de finalizacao do processo de limpeza
    print(str(datetime.now()) + ': Organizacao dos arquivos finalizada.')
```

```python
def main():
    """
    Organiza os dados meteorologicos das estacoes automaticas. 
    Origem: Instituto Nacional de Meteorologia 
    """
    
    ### Organiza os arquivos
    organize_files()

if __name__ == '__main__':
    main()
```
